```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

from sklearn.metrics import accuracy_score
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report

from concrete.ml.sklearn.rf import RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier as skRF

import time
import pandas as pd
import numpy as np
from numpy import mean
from numpy import std
```

```python
# Reading the dataset
dataset = pd.read_csv("gdrive/MyDrive/Special-Problem-Colab/Brain_GSE50161.csv")
dataset.head()
```

```python
feature_cols = [c for c in dataset.columns[2:]]
x = dataset.loc[:,feature_cols].values #must be floats
y = dataset.loc[:,'type'].values #must be integers
```

```python
# Preprocessing with labels for the lineage
le = preprocessing.LabelEncoder()
y = le.fit_transform(y)
le_mapping = dict(zip(le.classes_, range(len(le.classes_))))
print(le_mapping)
# print(y)

# x = x.astype(float)
# print(x)
# print("done!", x.shape)
```

```python
# Feature Selection

from sklearn.feature_selection import SelectKBest, chi2

print("\nUsing K best features feature selection...")
print("Shape of x before selection: ", x.shape)
selector = SelectKBest(chi2, k = 20)
x_new = selector.fit_transform(x, y)
x = x_new
print("Shape of x after selection: ", x.shape)
print("\n", x)
```

```python
# Get most important features accorting to Kbest

cols_index = selector.get_support(indices=True)
most_important_features = []

print("\nSelected features: ")
for col in cols_index:
  most_important_features.append(str(feature_cols[col]))
print(most_important_features)

print("\n")
print(dataset[most_important_features])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.10)

print(f"training set size: {X_train.shape[0]}")
print(f"testing set size: {X_test.shape[0]}")
```

```python
# RandomForest
skmodel = skRF(class_weight='balanced')
start_time = time.time()
skmodel.fit(X_train,y_train)
print(f"Running time for sklearn training is {time.time() - start_time} seconds")
start_time = time.time()
y_pred_clear = skmodel.predict(X_test)
print(f"Running time for sklearn prediction is {time.time() - start_time} seconds")


quant = RandomForestClassifier(class_weight='balanced')
start_time = time.time()
quant.fit(X_train, y_train)
print(f"Running time for quantized plaintext training is {time.time() - start_time} seconds")
start_time = time.time()
y_pred_q = quant.predict(X_test)
print(f"Running time for quantized plaintext prediction is {time.time() - start_time} seconds")

print("\n Random Forest Results \n")

# Accuracy
sk_accuracy = accuracy_score(y_test, y_pred_clear) * 100
quant_accuracy = accuracy_score(y_test, y_pred_q) * 100
print(f"Sklearn accuracy: {sk_accuracy:.4f}")
print(f"Quantized Clear Accuracy: {quant_accuracy:.4f}")

# Balanced Accuracy
sk_bal_accuracy = balanced_accuracy_score(y_test, y_pred_clear) * 100
quant_bal_accuracy = balanced_accuracy_score(y_test, y_pred_q) * 100
print(f"Sklearn Balanced accuracy: {sk_bal_accuracy:.4f}")
print(f"Quantized Clear Balanced Accuracy: {quant_bal_accuracy:.4f}")

# F1 Score
sk_f1 = f1_score(y_test, y_pred_clear, average='weighted') * 100
quant_f1 = f1_score(y_test, y_pred_q, average='weighted') * 100
print(f"Sklearn F1 Score: {sk_f1:.4f}")
print(f"Quantized Clear F1 Score: {quant_f1:.4f}")

# Recall / Sensitivity Score
sk_recall = recall_score(y_test, y_pred_clear, average='weighted') * 100
quant_recall = recall_score(y_test, y_pred_q, average='weighted') * 100
print(f"Sklearn Recall Score: {sk_recall:.4f}")
print(f"Quantized Clear Recall Score: {quant_recall:.4f}")

print("\n")
print("SKLEARN PREDICTION:\n", y_pred_clear)
print("QUANTIZED CLEAR PREDICTION:\n", y_pred_q)
print("ACTUAL:\n", y_test)
```

```python
# Random Forest FHE Model
start_time = time.time()
fhe = quant.compile(x)
print(f"Running time for FHE compilation is {time.time() - start_time} seconds")
start_time = time.time()
y_pred_fhe = quant.predict(X_test, fhe='execute')
print(f"Running time for FHE prediction is {time.time() - start_time} seconds")

# Results for FHE Models

print("FHE Random Forest Results \n")
# Accuracy
fhe_accuracy = accuracy_score(y_test, y_pred_fhe) * 100
print(f"Accuracy: {fhe_accuracy:.4f}")
# Balanced Accuracy
fhe_bal_accuracy = balanced_accuracy_score(y_test, y_pred_fhe) * 100
print(f"Balanced accuracy: {fhe_bal_accuracy:.4f}")
# F1 Score
fhe_f1 = f1_score(y_test, y_pred_fhe, average='weighted') * 100
print(f"F1 Score: {fhe_f1:.4f}")
# Recall / Sensitivity Score
fhe_recall = recall_score(y_test, y_pred_fhe, average='weighted') * 100
print(f"FHE Recall Score: {fhe_recall:.4f}")

print("\n")
print("SKLEARN PREDICTION:\n", y_pred_clear)
print("QUANTIZED CLEAR PREDICTION:\n", y_pred_q)
print("FHE PREDICTION:\n", y_pred_fhe)
print("ACTUAL:\n", y_test)

print("\n")
print(f"Quantized vs FHE Comparison: {int((y_pred_fhe == y_pred_q).sum()/len(y_pred_fhe)*100)}% similar")
print(f"Sklearn vs FHE Comparison: {int((y_pred_fhe == y_pred_clear).sum()/len(y_pred_fhe)*100)}% similar")

```
